### p54

```python
fit3=pd.read_csv("C:/Users/USER/Desktop/포스코 청년 아카데미_B4_김민경/데이터 분석1/FITNESS.CSV")
fit3
```

```python
# 결측치가 있는 1행 GENDER 부분에 성별 채워줌
fit3['GENDER'].iloc[1]='F'
# 23행의  AGE 값이  44이므로 AGEGROUP을 40으로 바꿔줌
fit3['AGEGROUP'].iloc[23]=40
```

```python
# 변경된 결과 확인
fit3
```

```python
fit3_group = fit3[['GENDER','AGEGROUP','WEIGHT']]
fit3_group
```

```python
# GENDER, AGEGROUP 로 그룹화 시킨 후 평균을 구한다. 
df_group = fit3_group.groupby(['GENDER','AGEGROUP']).agg('mean')
df_group
```

```python
# 행 인덱스를 초기화 시킨다. 
df_group.reset_index(inplace = True)
df_group
```

```python
# 'F'값을 찾아서 '여성'으로 바꾸고 'M'값을 찾아서 '남성'으로 변경합니다.
df_group.replace('F','여성', inplace = True)
df_group.replace('M','남성', inplace = True)
df_group
```

```python
# AGEGROUP 에서'30'값을 찾아서 '30대'으로 변경하고
# AGEGROUP 에서'40'값을 찾아서 '40대'으로 변경하고
# AGEGROUP 에서'50'값을 찾아서 '50대'으로 변경합니다.
df_group.replace({'AGEGROUP':{30:'30대'}},inplace = True)
df_group.replace({'AGEGROUP':{40:'40대'}},inplace = True)
df_group.replace({'AGEGROUP':{50:'50대'}},inplace = True)
df_group
```

```python
# index를 GENDER로 columns을 AGEGROUP로 하고, 값에는 평균으로 채워준다.
df_group1 = df_group.pivot(index = 'GENDER', columns = 'AGEGROUP', values = 'WEIGHT')
df_group1
```
