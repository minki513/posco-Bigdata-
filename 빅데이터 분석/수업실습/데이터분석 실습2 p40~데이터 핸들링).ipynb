### p40

```python
import pandas as pd
import numpy as np
from numpy.random import randn  
```

```python
left = pd.DataFrame({
    "key1": [1, 2, 3, 4],
    "A": ["A0", "A1", "A2", "A3"],
    "B": ["B0", "B1", "B2", "B3"]})

right = pd.DataFrame({
    "C": ["C1", "C2", "C5"],
    "D": ["D1", "D2", "D5"]},
    index=[1, 2, 5])
display(left, right)
```

```python
# index 기준으로
result = left.join(right)
result
```

```python
# right 를 기준으로 변경하라
result = left.join(right, how = 'right')
result
```

```python
# 인덱스 번호가 같은 것끼리
result = left.join(right, how = 'inner')
result
```

```python
# 모든게 다나옴
result = left.join(right, how = 'outer')
result
```

```python
# key1 을 기준으로 결합하라
result = left.join(right, on = 'key1')
result
```

```python
result2 = pd.merge(left, right, left_on = 'key1', right_index = True, how = 'left')
result2
```

### p50

```python
df = pd.DataFrame({
    "index": [1,1,1,2,2,2], # 남 남 남 여 여 여
    "column": ["col_1", "col_2", "col_3", "col_1", "col_2", "col_3"], #10대, 20대~
    "value_1": [1, 2, 3, 4, 5, 6],  
    "value_2": ["x", "y", "z", "q", "w", "t"]})
df
```

```python
# index 와 columns설정해줌, 값은 value_1으로 넣어줌
df.pivot(index = 'index', columns = 'column', values = 'value_1')
```

```python
# 위에 코드랑 같은 방식
df.pivot(index = 'index', columns = 'column')['value_1']
```

```python
# 복수의 pivot column 지정
df.pivot(index = 'index', columns = 'column', values =['value_1','value_2'])
```

```python
# 미지정하면 그냥 모든 값을 출력
df.pivot(index = 'index', columns = 'column')
```

```python
# index 를 미지정한다면
df.pivot(columns = 'column', values = 'value_1')
```

```python
# melt 이용하여 변환 풀어서 쌓음
df.melt(id_vars = ['index','column'])
```

```python
# value_1만 쌓음
df.melt(id_vars = ['index','column'], value_vars = ['value_1'])
```

```python
df.melt(id_vars = ['index','column'], value_vars = ['value_2'])
```

```python
df = pd.DataFrame({
    "index": [1,2,3],
    "alpha": ['A','B','C'], #10대, 20대~
    "values": [11, 22, 33],  
    "xyz": ["x", "y", "z"]})
df
```

### p62

```python
df = pd.DataFrame({
        "factor1": ['A','A','A','A','A','B','B','B','B'],
        "factor2": ['a','b','c','a','b','a','b','c','c'],
        "value": [11, 13, 12, 12, 14, 12, 13, 11, 12]})
df
```

```python
pd.crosstab(df.factor1, df.factor2)
```

```python
pd.crosstab(df.factor1, df.factor2, rownames = ['Fact1'],colnames = ['Fact2'],
           margins = True)
```

```python
pd.crosstab(df.factor1, [df.factor2,df.value])
```

```python
pd.crosstab(df.factor1, df.factor2, margins = True, normalize = True)
```
