### p21

```python
df = pd.DataFrame({"ABC": ["A","B","C","A"], 
    "abc":["a","b","c","c"],
    "group1":["G1","G2","G3","G4"],
    "value1":[11,13,12,14],
    "val1_mod":[0.231,0.197,0.213, 0.199] })
df
```

```python
# 컬럼이름으로 가져오기
df.filter(items = ['group1', 'value1'])
```

```python
# regex : 입력 값을 포함한 컬럼 선택. '1'포함해서 가져오기 (조건주기) axis = 1 열을 기준으로 가져와라
df.filter(regex = '1', axis = 1)
```

```python
# 1로 끝나는 컬럼
df.filter(regex = '1$', axis = 1)
```

```python
# a 포함된 컬럼
df.filter(like = 'a', axis = 1)
```

```python
# A 또는 a가 포함된 컬럼
df.filter(regex = '[Aa]', axis = 1)
```

```python
# value1 이 12보다 큰 값을 가린다.
df.mask(df.value1 > 12)
```

```python
df.mask(df.abc == 'a')
```

```python
# 조건에 맞는 곳에 999 대입
df.mask(cond = df.value1 > 12, other = 999)
```

### p28

```python
df1 = pd.DataFrame({
        "A": ["A0", "A1", "A2", "A3"],
        "B": ["B0", "B1", "B2", "B3"],
        "C": ["C0", "C1", "C2", "C3"],
        "D": ["D0", "D1", "D2", "D3"]},
        index=[0, 1, 2, 3])

df2 = pd.DataFrame({
        "A": ["A4", "A5", "A6", "A7"],
        "B": ["B4", "B5", "B6", "B7"],
        "C": ["C4", "C5", "C6", "C7"],
        "D": ["D4", "D5", "D6", "D7"]},
        index=[4, 5, 6, 7])

df3 = pd.DataFrame({
        "A": ["A8", "A9", "A10", "A11"],
        "B": ["B8", "B9", "B10", "B11"],
        "C": ["C8", "C9", "C10", "C11"],
        "D": ["D8", "D9", "D10", "D11"]},
        index=[8, 9, 10, 11])
display(df1, df2, df3)
```

```python
# df1, df2, df3, 결합
result = pd.concat([df1, df2, df3])
result
```

```python
# append 이용해서 추가해도 똑같음
df1.append([df2, df3])
```

```python
result = pd.concat([df1, df2, df3], axis = 1)
result
```

```python
# ignore_index 를 true하면 원래 열의 인덱스를 무시하고 0부터 차례대로 재편성함
result = pd.concat([df1, df2, df3], axis = 1, ignore_index = True)
result.columns = ['col' + str(i) for i in range(12)]
result
```

```python
# ignore_index 를 true하면 원래 행의 인덱스를 무시하고 0부터 차례대로 재편성함
result = pd.concat([df1, df2, df3], axis = 0, ignore_index = True)
result
```

### p36

```python
left = pd.DataFrame({
        "key1": ["K0", "K1", "K2", "K3"],
        "key2": ["K0", "K1", "K0", "K1"],
        "A": ["A0", "A1", "A2", "A3"],
        "B": ["B0", "B1", "B2", "B3"]})

right = pd.DataFrame({
        "key1": ["K0", "K1", "K2", "K3"],
        "key2": ["K0", "K0", "K0", "K0"],
        "C": ["C0", "C1", "C2", "C3"],
        "D": ["D0", "D1", "D2", "D3"]})
display(left, right)
```

```python
#'key1', 'key2'을 결합하라 - 같은거 끼리만 결합된다.
result = pd.merge(left, right, on = ['key1', 'key2'])
result
```

```python
#key1', 'key2'을 결합하는데 왼쪽을 기준으로 붙여라
result = pd.merge(left, right, on = ['key1', 'key2'], how = 'left')
result
```

```python
#key1', 'key2'을 결합하는데 오른쪽을 기준으로 붙여라
result = pd.merge(left, right, on = ['key1', 'key2'], how = 'right')
result
```

```python
#key1', 'key2'을 결합하는데 같지 않아도 다 나오게 한다.
result = pd.merge(left, right, on = ['key1', 'key2'], how = 'outer')
result
```

```python
# indicator 기준치를 표시해줌
result = pd.merge(left, right, on = ['key1', 'key2'], how = 'outer', indicator = True)
result
```
