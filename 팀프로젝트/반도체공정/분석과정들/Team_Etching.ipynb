```python
def Etching_rate(df_raw):
    return (df_raw["Thin F1"] - df_raw["Thin F4"]) / 40
df_raw["Etching_rate(nm/min)"] = df_raw.apply(Etching_rate, axis =1)

def Etching_rate1(df_raw):
    return (df_raw["Thin F1"] - df_raw["Thin F2"]) / 40
df_raw["Etching_rate1(nm/min)"] = df_raw.apply(Etching_rate1, axis =1)

def Etching_rate2(df_raw):
    return (df_raw["Thin F2"] - df_raw["Thin F3"]) / 40
df_raw["Etching_rate2(nm/min)"] = df_raw.apply(Etching_rate2, axis =1)

def Etching_rate3(df_raw):
    return (df_raw["Thin F3"] - df_raw["Thin F4"]) / 40
df_raw["Etching_rate3(nm/min)"] = df_raw.apply(Etching_rate3, axis =1)

df_raw
```

```python
df_raw[["Etching_rate(nm/min)", "Etching_rate1(nm/min)", "Etching_rate2(nm/min)","Etching_rate3(nm/min)"]].hist(bins = 50, figsize = (15,5))
```

```python
df_mean = df_raw[["Etching_rate(nm/min)"]].mean()
df_max = df_raw[["Etching_rate(nm/min)"]].max()
df_min = df_raw[["Etching_rate(nm/min)"]].min()

def Non_Uniformity(df_raw):
    return (df_max - df_min) / (2*df_mean)

df_raw["Non_Uniformity"] = df_raw.apply(Non_Uniformity, axis =1)
df_raw
```
