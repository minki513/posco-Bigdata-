```python
class SET:
    
    def __init__(self, ss=[]):   # 생성자 
        result = []  # 결과를 담을 빈 리스트
        self.ss = ss  # 들어오는 리스트 
        #print(self.ss)
        for i in self.ss:  
            if i not in result:  # 해당리스트에 값이 결과리스트에 없는 것만 = 중복제거
                result.append(i)# 추가한다
        result.sort()  # 오름차순으로 정렬 
        self.ss= result # 그 결과값 self.ss에 넣기
 
     
    def __str__(self):  # {} 형태로 반환 print 기능
        temp=''
        for i in range(len(self.ss)):  
            if i == (len(self.ss)-1):  #만약 i가 (리스트의 길이-1)= 마지막 인덱싱 값과 같다면
                temp += str(self.ss[i])  # 마지막으로 추가
            else:
                temp += str(self.ss[i]) +', '  #  리스트의 값을 하나씩 차례대로 채워줌
        return '{'+ temp +'}'  # 마지막으로 {} 이안에 값을 담아 출력

    
    def add(self, elem):  # 추가하는 연산 
        if elem not in self.ss:  # 이 값이 현재 결과값리스트에 없다면 
            return  self.ss.append(elem)  # 리스트의 append를 이용하여 추가하라
        return self.ss   # 이미 있는 값이면 그냥 원래 결과리스트 출력
    
    
    def discard(self, elem):  # 제거하는 연산
        if elem in self.ss:  # 이 값이 현재 결과리스트에 있다면 
            return  self.ss.remove(elem)   # 그 값을 지워라 
        return self.ss   # 이미 있는 값이면 그냥 원래 결과리스트 출력
    
    
    def clear(self):   #원소 제거하는 함수
        return self.ss.clear()  # clear함수사용
    
    def __len__(self):   #길이 계산
        return len(self.ss) 
    
    def __contains__(self,elem):  # 값의 존재 유무
        for i in range(len(self.ss)):  # 결과값의 길이만큼 차례대로 for문 진행
            if elem in self.ss:  # 들어온 값이 그 결과값 리스트에 있으면
                return True  # true 출력
            return False

    def __or__(self,other): # 합집합
        dd=[]  # 빈 리스트 생성
        for i in other.ss:  # 오른쪽 리스트에서 하나씩 뽑아
            if i not in self.ss :    # 왼쪽 리스트와 비교 없으면
                dd.append(i)    # 빈 리스트에 추가 - 두 리스트중에 겹치지 않는 값이 들어감
        for i in range(len(self.ss)):  
            dd.append(self.ss[i])  # 왼쪽 리스트를 인덱싱해서 하나씩 dd 리스트에 담아줌
            dd.sort()
        
        return SET(dd)             

    def __and__(self,other): # 교집합 
        dd= []  # 빈리스트 생성
        for i in self.ss:  # 왼쪽 리스트의 값들을 하나씩
            if i in other.ss:    # 오른쪽 리스트와 비교하여 같은 값이 있다면
                dd.append(i) # dd리스트에 담아준다.
                dd.sort()
        return SET(dd)
       

    def __sub__(self,other):  # 차집합
        dd=[]   # 빈리스트 생성
        for i in self.ss:   # 왼쪽 리스트값들을
            if i not in other.ss:  # 오른쪽 리스트에 없다면 
                dd.append(i)   # dd 리스트에 추가
                dd.sort()
        return SET(dd)
                
       
    def __le__(self, other):  # self <= other 부분집합
        for i in self.ss:   # 왼쪽 리스트에서 하나씩 뽑는다.
            if i not in other.ss:  # 오른쪽 리스트에 그 값이 없다면
                return False   # 부분집합이 아니다
        return True  
    
    def __ge__(self, other):  # self >= other 부분집합
        for i in other.ss:
            if i not in self.ss:
                return False
        return True
           
        
    def __iand__(self,other):  # &=   교집합과 돌아가는 것이 같다.
        dd= []
        for i in other.ss:  
            if i in self.ss:  
                dd.append(i)
                self.ss=dd
            return self
    
    
    def __ior__(self,other):  # |=   합집합과 돌아가는 것이 같다.
        for i in other.ss:  
            if i not in self.ss :  
                self.ss.append(i)
        return self
    
       
    def __isub__(self, other):  # -=   차집합과 돌아가는 것이 같다
        for i in other.ss:
            if i in self.ss:
                self.ss.remove(i)
        return self
```

```python
a = SET([1,2,3,4])
b = SET([1,2,3,4])
```

```python
print(a)
print(b)
print()

a.discard(4)
b.discard(1)
print(a)
print(b)
print()

print(len(a))
print(1 in a)
print(1 in b)
print()

print(a | b)
print(a & b)
print(a - b)
print()

print(a <=b)
print(a <= a | b) 
print(a >= b)
print(a >= a & b)

b.clear()
print(b)
print()
```
